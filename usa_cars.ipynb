### 1. top 5 car brand in terms of number of cars

```python
data=[]

for brand in df['brand'].unique():
    data.append([brand,len(df[df['brand']==brand])])

pd.DataFrame(data,columns=['brand','freq']).sort_values(by='freq',ascending=False).head().plot(x='brand',y='freq',kind='bar',figsize=(10,5))
```

### 2.car brand based on frequency of cars

```python
data=[]

for brand in df['brand'].unique():
    data.append([brand,len(df[df['brand']==brand])])

pd.DataFrame(data,columns=['brand','freq']).sort_values(by='freq',ascending=False).plot(x='brand',y='freq',kind='bar',figsize=(10,5))
```

### 3.car industry trend

```python
data=[]

for year in df['year'].unique():
    data.append([year,len(df[df['year']==year])])

pd.DataFrame(data,columns=['year','freq']).sort_values(by='year',ascending=True).plot(x='year',y='freq',kind='line',figsize=(10,5))
```

## 4.most popular car colours

```python
data=[]

for brand in df['color'].unique():
    data.append([brand,len(df[df['color']==brand])])

pd.DataFrame(data,columns=['color','freq']).sort_values(by='freq',ascending=False).plot(x='color',y='freq',kind='bar',figsize=(10,5))
```

### 5.which car brand is having maximum average price

```python
df.groupby('brand').mean()['price'].sort_values(ascending=False).plot.bar()
```

### 6.which state is having most expensive cars

```python
df.groupby('state').mean()['price'].sort_values(ascending=False).plot.bar(figsize=(15,5))
```

##7.which year most expensive car manufactured

```python
df.groupby('year').mean()['price'].sort_values(ascending=False).plot.bar(figsize=(15,5))
```
